```python
import gc
import json
import boto3
import decimal
import pandas as pd
import numpy as np
```

```python
class DecimalEncoder(json.JSONEncoder):
    def default(self, i):
        if isinstance(i, decimal.Decimal):
            if i%1>0:
                return float(i)
            else:
                return int(i)
        return super(DecimalEncoder, self).default(i)
    
```
