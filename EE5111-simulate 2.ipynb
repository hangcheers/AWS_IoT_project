```python
# connection setting
SHADOW_CLIENT = "A0195009A_ShadowClient1"
HOST_NAME = "a2ebf8r9itm3yg-ats.iot.us-west-2.amazonaws.com"
ROOT_CA = "/Users/helena/Documents/EE5111/AmazonRootCA1.pem"
PRIVATE_KEY = "/Users/helena/Documents/EE5111/3d84d4e6ce-private.pem.key"
CERT_FILE = "/Users/helena/Documents/EE5111/3d84d4e6ce-certificate.pem.crt"
SHADOW_HANDLER = "A0195009A1"
```

```python
# automatically call whenever the shadow is updated
def myShadowUpdateCallback(payload, responseStatus, token):
#     print()
#     print('UPDATE: $aws/things' + SHADOW_HANDLER + '/shadow/update/#')
#    print("payload = " + payload)
    print("respnseStatus =" + responseStatus)
#     print("token =" + token)

# create, configure, and connect a shadow client
myShadowClient = AWSIoTMQTTShadowClient(SHADOW_CLIENT)
myShadowClient.configureEndpoint(HOST_NAME, 8883)
myShadowClient.configureCredentials(ROOT_CA, PRIVATE_KEY, CERT_FILE)
myShadowClient.configureConnectDisconnectTimeout(10)
myShadowClient.configureMQTTOperationTimeout(5)
myShadowClient.connect()

myDeviceShadow = myShadowClient.createShadowHandlerWithName(SHADOW_HANDLER, True)
```

```python
#simulate 2nd engine 
print('Program Start')
trainDB = pd.read_csv('/Users/helena/Documents/EE5111/train_FD002.txt',
                      delim_whitespace=True, header=None)
sensor_name = ['s' + str(i) for i in range(1, 22)]
column_list = ['id', 'te', 'os1', 'os2', 'os3'] + sensor_name
df = pd.DataFrame(trainDB.values, columns=column_list)
# overwrite column'id' of the engine as 'FD001'+id
df['id'] = df['id'].map(lambda s: 'FD002_' + str(s))
# add one more column 'timestamp'
# add one more column that contains matric number
df['matric number'] = 'A0195009A'
for i in range(5000):
    print("now publishing no." + str(i + 1) + "row")
    row = df.loc[[i]]
    row['timestamp'] = 'UTC' + str(datetime.utcnow())
    row.index = ["reported"]
    row = row.to_json(orient='index')
    row = '{"state": ' + row + '}'
    myDeviceShadow.shadowUpdate(row, myShadowUpdateCallback, 5)
    time.sleep(2)
```
